We got archives files in **.gz** type, we need to extract their content, ket's see an example of reading one of them :

```python
import gzip
import os
files = []

from Bio.Affy import CelFile

def read_cel(file):
    c = CelFile.read(file)
    print("INTENSITIES")
    print(c.intensities)
    print("STDEVS")
    print(c.stdevs)
    print("NPIX")
    print(c.npix)
    

def read_gz(file):
    with gzip.open(file,'r') as fin:        
        read_cel(fin)      

        
print(str(onlyfiles[1]))    
read_gz(os.path.join(mypath, onlyfiles[1]))
```
