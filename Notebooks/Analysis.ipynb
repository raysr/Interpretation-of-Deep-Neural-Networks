```python
import gzip
import os
files = []

from Bio.Affy import CelFile

def read_cel(file):
    c = CelFile.read(file)
    print(help(c))
    for i in c.__dict__.keys():
        print(str(i))
        print(str(c.__dict__[i]))
        print("\n")
def read_gz(file):
    with gzip.open(file,'rb') as fin:        
        read_cel(fin)      
    
        
#print(str(onlyfiles[0]))   
read_gz(os.path.join(mypath, onlyfiles[0]))
```
